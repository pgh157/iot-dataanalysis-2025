````python
# 기본데이터
arr = np.random.random(50) * 100
arr
````

````python
# 단일차트(라인차트)
plt.plot(arr)
````

````python
# 출력메시지를 삭제하려면
plt.plot(arr)
plt.show()
````

````python
# 분포표(정규분포)
plt.hist(np.random.randn(100))
plt.show()
````

````python
# 산점도
plt.scatter(np.arange(30), np.random.randn(30))
plt.show()
````

````python
# 다중차트 1번째 - 여러차트를 한꺼번에 출력
fig = plt.figure(figsize= (15,8))      # 객체 생성

ax1 = fig.add_subplot(2, 2, 1)          # 2x2 차트 중 1번째
ax2 = fig.add_subplot(2, 2, 2)          # 2x2 차트 중 2번째
ax4 = fig.add_subplot(2, 2, 4)          # 2x2 차트 중 4번째

ax1.hist(np.random.randn(100))
ax2.scatter(np.arange(30), np.random.randn(30))
ax4.plot(arr)
plt.show()
````

````python
# 다중차트 2번째 - 여러차트를 한꺼번에 출력
# 차트꾸미기
fig, ax = plt.subplots(2,2, figsize =(15,8))

# color -> 색상변경, marker 포인트모양 (o, ^, *, 1)
# linestyle -> -, --, :, -. 
ax[0][0].hist(np.random.randn(100), color = 'green')
ax[0][1].scatter(np.arange(30), np.random.randn(30), color = 'blue', marker = '1')
ax[1][0].plot(arr, color='crimson',linestyle = '-', marker= 'o')
ax[1][1].plot(arr, color='violet',linestyle = '-.', marker= 'o')

plt.show()
````

````python
# 모듈 사용
import folium
import json
from folium import Marker

folium.__version__
````

````python
# 서울 중심 위치, 위경도는 소수점 4자리 이상으로 사용해야 오차적음
loc = (37.5776087830657, 126.976896737645)  #  경복궁

json_data = open('./seoul_municipalities_geo_simple.json', encoding= 'utf-8').read()
data = json.loads(json_data)


m = folium.Map(location = loc, zoom_start = 11)         # 줌 0 ~ 19
folium.GeoJson(data, name = '동').add_to(m)
Marker(location= loc, popup='경복궁', icon = folium.Icon(color='lightblue', icon='heart')).add_to(m)
m

# folium.Icon 종류 - https://icons.getbootstrap.com/ 검색
````
